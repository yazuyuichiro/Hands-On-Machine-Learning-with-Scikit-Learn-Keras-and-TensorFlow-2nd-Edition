### エンコーダーデコーダーネットワーク（系列→系列）図16−3、図16-4を確認
処理の方針は人間の脳に非常に近い。\
I drink milk.を単語ごとに分割し、順番に読み、意味を理解する。その意味になるようにフランス語から単語を選択する。


<u>・エンコーダー（系列→意味）</u>
- 入力として、文章を受け取って文章の意味ベクトルを出力する。
- 入力を反対にするほうが都合が良い（？）

**隠れ状態の動き**\
まず始まる前は、これから最初の単語が入ってくるんだなと理解している\
Iという単語が入ってきて、私に関する文章だと理解する\
drinkが入ってきて、私は何かを飲むという意味を表すと理解する\
milkが入ってきて、私は牛乳を飲むという文章を理解する\
最後にピリオドが入ってきて終了だとわかる\
という情報を意味ベクトルとして出力
<br>
<br>

<u>・デコーダー（意味→系列）</u>（時系列の意識をするとわかりやすい）
- 意味ベクトルを入力すると、意味ベクトルの数値とデコーダーで直前に出力した単語の情報から今出力する単語の候補をスコアで出力→ソフトマックスで確率にし高いものを出力とする
  - １ステップシフトバックするというのは、学習時の話だと理解している。正解データを与えているという認識。
<br>
<br>

<u>・単語埋め込み</u>
- 単語の特徴をベクトルで表現したもの（200次元でうまく表現：例えば王様というベクトルから男性というベクトルを引き、最後に女性というベクトルをたすと女王というベクトルが生成される）
<br>
<br>

### 注意点
- ベクトルは固定長なので、扱える分の長さや複雑さに制限ある
  - 入出力は短い文章にパディングすることで長さを揃える。下記で実装可能。
```python
tf.data.experimental.bucket_by_sequence_length()
```
- EOSトークンが入れば文章は終了ですよ
- 間違った単語のサンプルと正しい単語のサンプルのロジット（オッズのようなもの？）を比較して、計算コストを下げる
  - そうしないと、英語からフランス語に翻訳する場合、フランス語の単語のすべてが選択肢となってしまう

### コードの解説

````python
# 前範囲より
vocab_size = 10000
embed_size = 128
````

````python
import keras
import tensorflow as tf
import numpy as np
import tensorflow_addons as tfa

encoder_inputs = keras.layers.Input(shape=[None], dtype=np.int32)
decoder_inputs = keras.layers.Input(shape=[None], dtype=np.int32)
sequence_lengths = keras.layers.Input(shape=[], dtype=np.int32)

# embedding：自然言語を計算が可能な形に変換すること
embeddings = keras.layers.Embedding(vocab_size, embed_size)
encoder_embeddings = embeddings(encoder_inputs)
decoder_embeddings = embeddings(decoder_inputs)

# return state=Trueで隠れ状態の最後の状態（すべての単語を処理した状態、文章の意味を理解した状態）を返す
encoder = keras.layers.LSTM(512, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_embeddings)
encoder_state = [state_h, state_c]

# 自身の出力を使うのか正解データから使うのか
sampler = tfa.seq2seq.sampler.TrainingSampler()

decoder_cell = keras.layers.LSTMCell(512)
output_layer = keras.layers.Dense(vocab_size)
decoder = tfa.seq2seq.basic_decoder.BasicDecoder(decoder_cell, sampler,
                                                 output_layer=output_layer)
final_outputs, final_state, final_sequence_lengths = decoder(
    decoder_embeddings, initial_state=encoder_state,
    sequence_length=sequence_lengths)
Y_proba = tf.nn.softmax(final_outputs.rnn_output)

model = keras.models.Model(
    inputs=[encoder_inputs, decoder_inputs, sequence_lengths],
    outputs=[Y_proba])
````

````python
model.summary()
````

## 16.3.1 双方向RNN
時系列に関しては、過去の情報を入力、未来の予測を出力\
文章に関しては、（英語の文章を逆から翻訳した方がやりやすいみたいな受験のときしか使えないようなテクニックがあったように）文章の先頭から順番に読むだけでなく、後ろにある情報も翻訳には必要\
前から読んだ情報の与え方と後ろから読んだ情報の与え方の２種類で意味ベクトルを生成する方法

例：イギリスの女王なのかハートのクイーンなのか、女王蜂なのかを意味ベクトルに含めるためには女王という単語の前後の単語を理解する必要がある

## 16.3.2 ビームサーチ
（デコーダーの話）\
問題：入力単語（直前の情報）から可能性が高い単語を出力とした。完成した文章の全体を見た時に違和感のある出力となった。\
ビームサーチとは、k(任意の数)個の選択肢を保管し続ける手法。

#### ステップ1
How(75%)\
What(3%)\
You(1%)\
<br>

#### ステップ2
How(75%)\
->will(36%) : `27%`\
->are(32%) : `24%`\
->do(16%) : `12%`\
<br>
What(3%)\
->do(50%) : `1.5%`\
->are(7%) : `0.22%`\
-> ...\
<br>
You(1%)\
-> ...\
<br>

#### ステップ3
How will(27%)\
-> you(8%)：`2.1%`\
-> ...\
<br>
How are(24%)\
-> you(42%)：`10.1%` \
-> ...


## 16.4 注意機構（Attention機構）
エンコーダーデコーダーモデルでは、意味ベクトルの次元が固定→表せる意味や複雑さに限界がある（限界を超えると精度が下がる）\
注意機構では、前モデル同様に固定次元の意味ベクトルと直前の出力の情報を利用。それに加えて注意機構では入力単語への参照を情報として利用する\
入力単語の参照とは入力された文章のそれぞれの単語ベクトルに、文章で合計が1になるように重みを割り振ったもの（注目したい単語だけを集める）\
より人っぽく処理している。\
重みの取り方として2種類掲載されている
- Bahdanau注意
- Luong注意

<u>エンコーダー</u>
- 入力として、単語を受け取って特徴ベクトルを出力する。
- 前後の文脈を加味した単語ベクトル（計算方法に双方向RNNを使用）
<br>
<br>

<u>デコーダー</u>
- 意味ベクトルを受け取ると、「出力した直前の単語、文章の意味、エンコーダーの入力文（翻訳前の文）の重み」という情報から出力したい単語の確率を計算する
<br>
<br>
<u>アライメントモデル</u>
- 2層のNNを使って、いま時点の意味ベクトルと入力単語ベクトル全部の情報からどの単語に注目するかを重みで表現する
