具体的に本の例でいくと、カラー画像に5✕5のカーネルを使って特徴量マップを200個出力する場合

```python
tensorflow.keras.layers.Conv2D(200, (5, 5), activation="relu", input_shape=(150,100,3))

カーネルの計算だけで\
5 ✕ 5 ✕ 3 ✕ 200 = 15,000\
の特徴があることになる。\
このカーネルを使って入力画像(150✕100)の計算をしていくので、あとは単純な計算ではあるが億単位に余裕で到達する。

さらに\
その計算が32ビット浮動小数点(float32)だったら、、、　→16ビット浮動小数点(float16)に変換\
1バッチに入力画像が100枚で設定していたら、、、　→バッチサイズを小さくする

特に訓練中は逆伝播のためにすべてのパラメータを残しておく必要があるのでRAMの負荷が多くなる。

## 14.3 プーリング層

P452\
**プーリング層の目的：入力画像のサブサンプリング(縮小)**\
→重要な特徴を残しながら、計算負荷・メモリの使用量・パラメータ数の削減、過学習の緩和を行う

プーリング層には重みはなく、決められたルールで計算するだけ。

**ルール**
- プーリングカーネルのサイズ
- ストライド
- 集計関数(最大値を伝えるのか、平均値を伝えるのか)
- パディングの有無(無しの場合、端数は切り捨て)

**◆具体的にどのように計算するか**\
カーネルサイズ(2✕2)、ストライド2、マックスプーリングのとき

このようにフィルタを通していき、その中の最大値や平均値を新たな特徴として伝える。\
これによって画像の特徴をぼんやりと把握するということができるようになる。\
ある1つの特徴に対して計算結果を返すだけなので、深さは変わらない。

## 14.4 CNNのアーキテクチャ

````python
# FashionMNISTの例
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(64, 7, activation='relu', padding='same', input_shape=[28, 28, 1]),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Conv2D(128, 3, activation='relu', padding='same'),
    tf.keras.layers.Conv2D(128, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Conv2D(256, 3, activation='relu', padding='same'),
    tf.keras.layers.Conv2D(256, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Accuracyは0.92を超える
````

### 14.4.1 LeNet-5
1998年に開発。手書き文字認識で広く使われる。

````python
lenet_5 = tf.keras.Sequential([
tf.keras.layers.Conv2D(6, kernel_size=(5, 5), strides=(1, 1), padding='same', activation='tanh', input_shape=[32, 32, 1]),
tf.keras.layers.AveragePooling2D((2, 2), strides=(2, 2)),
tf.keras.layers.Conv2D(16, kernel_size=(5, 5), strides=(1, 1), padding='valid', activation='tanh'),
tf.keras.layers.AveragePooling2D((2, 2), strides=(2, 2)),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(84, activation='tanh'),
tf.keras.layers.Dense(10, activation='softmax')
])
````

### 14.4.2 AlexNet
2012年のコンペティションで圧勝し話題となったモデル。\
基本はLeNet-5の構造に似ていて、時代とともに計算能力が向上した背景もあり大規模で深いモデルになっている。\
畳み込み層を重ねた最初の構造である。

````python
alexnet = tf.keras.Sequential([
tf.keras.layers.Conv2D(96, kernel_size=(11, 11), strides=(4, 4), padding='valid', activation='relu', input_shape=[224, 224, 3]),
tf.keras.layers.AveragePooling2D((3, 3), strides=(2, 2)),
tf.keras.layers.Conv2D(256, kernel_size=(5, 5), strides=(1, 1), padding='same', activation='relu'),
tf.keras.layers.AveragePooling2D((3, 3), strides=(2, 2)),
tf.keras.layers.Conv2D(384, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu'),
tf.keras.layers.Conv2D(384, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu'),
tf.keras.layers.Conv2D(256, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu'),
tf.keras.layers.AveragePooling2D((3, 3), strides=(2, 2)),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(4096, activation='relu'),
tf.keras.layers.Dropout(0.5),
tf.keras.layers.Dense(4096, activation='relu'),
tf.keras.layers.Dropout(0.5),
tf.keras.layers.Dense(1000, activation='softmax')
])
````

→過学習抑制のために**データ拡張**を行っている\

画像の回転、拡大・縮小、を行ってデータセットに追加することで\

**局所応答正規化(LRN:Local response normalization)**\
P461\
これは側方抑制という生物の脳の動きを模倣したもの\
- 本能的にその生物にとって心地よい選択を行ってしまうことを抑制すること。(側方抑制機能がないと、生物は同じ運命をたどり絶滅する。)
- たちどまって考える、つまり思考の余白を作ることで知能が向上する。
- 一旦冷静になる

話をLRNに戻すと、\
これによりConv層から出力された特徴マップが広い範囲の特徴量を探ることができ、結果的に汎化性能があがる。

```python
tensorflow.nn.local_response_normalization()
